1. Write a function named is_vowel. It should accept a string as input and use a regular expression to determine if the passed string is a vowel. While not explicity mentioned in the lesson, you can treat the result of re.search as a boolean value that indicates whether or not the regular expression matches the given string.



```python
def is_vowel(string):
    """
    Checks if a string is a vowel
    """
    #convert to lowercase
    string = string.lower()
    return re.search(r'[aeiou]', string) is not None

is_vowel('A')    
```

2. Write a function named is_valid_username that accepts a string as input. A valid username starts with a lowercase letter, and only consists of lowercase letters, numbers, or the _ character. It should also be no longer than 32 characters. The function should return either True or False depending on whether the passed string is a valid username.


    is_valid_username('aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa')

    False

    is_valid_username('codeup')

    True

    is_valid_username('Codeup')

    False

    is_valid_username('codeup123')

    True

    is_valid_username('1codeup')

```python
not re.search(r'[aeiou]', 'a')
```

```python
test_list =['aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa', 'codeup', 'Codeup', 'codeup123', '1codeup', 'codeup_123', 'codeup_123_']

def is_valid_username(string):
    """
    Checks if a string is a valid username
    starts with lowercase letter, contains only lowercase letters, numbers, and underscores
    no longer than 32 characters
    """
    if len(string) > 32:
        return False
    if not re.search(r'^[a-z]', string):
        return False
    if not re.search(r'[a-z/n_]', string):
        return False
    return True

for test in test_list:
    print(is_valid_username(test))
```

3. Write a regular expression to capture phone numbers. It should match all of the following:


(210) 867 5309
+1 210.867.5309
867-5309
210-867-5309


```python
def phone_number(string):
    """
    Convert string to 10 digit phone number
    """
    #remove all non-numeric characters
    string = re.sub(r'\D', '', string)
    #check if string is 10 digits
    if len(string) == 10:
        return string
    elif len(string)==11 and string[0]=='1':
        return string[1:]
    else:
        return '210'+string

test_numbers = ['(210) 867 5309', '+1 210.867.5309', '867-5309', '210-867-5309']
for test in test_numbers:
    print(phone_number(test))
```

# Bonus Exercise

You can find a list of words on your mac at /usr/share/dict/words. Use this file to answer the following questions:


```python
word_list=pd.DataFrame()
# get list from /usr/share/dict/words
with open('/usr/share/dict/words', 'r') as f:
    word_list['word'] = f.read().splitlines()
# make all lowercase
word_list['word'] = word_list['word'].str.lower()
word_list
```



- How many words have at least 3 vowels?


```python
word_list['3_or_more_vowels'] = word_list['word'].apply(lambda x:(re.search(r'[aeiou].*[aeiou].*[aeiou]',x)) is not None)
word_list
```

- How many words have at least 3 vowels in a row?


```python
word_list['3_or_more_vowels_in_row'] = word_list['word'].apply(lambda x:(re.search(r'[aeiou]{3,}',x)) is not None)
word_list[word_list['3_or_more_vowels_in_row']==True]
```

- How many words have at least 4 consonants in a row?


```python
consonants = 'bcdfghjklmnpqrstvwxyz'
```

```python
word_list['4_or_more_consonants_in_row'] = word_list['word'].apply(lambda x:(re.search(r'[bcdfghjklmnpqrstvwxyz]{4,}',x)) is not None)
word_list[word_list['4_or_more_consonants_in_row']==True]
```

- How many words start and end with the same letter?


```python
word_list['start_and_end_same_letter'] = word_list['word'].apply(lambda x:x[0]==x[-1])
word_list[word_list['start_and_end_same_letter']==True]
```

- How many words start and end with a vowel?


```python
word_list['start_and_end_with_vowel'] = word_list['word'].apply(lambda x:is_vowel(x[0]) and is_vowel(x[-1]))
word_list[word_list['start_and_end_with_vowel']==True][['word','start_and_end_with_vowel']]
```

- How many words contain the same letter 3 times in a row?


```python
def same_letter_three_times(string):
    """
    Checks if a string has 3 or more consecutive letters
    """
    return re.search(r'(.)\1\1', string) is not None


word_list['same_letter_three_times'] = word_list['word'].apply(lambda x:same_letter_three_times(x))
word_list[word_list['same_letter_three_times']==True][['word','same_letter_three_times']]
```
